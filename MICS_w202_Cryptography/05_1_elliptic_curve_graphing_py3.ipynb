### Elliptic Curves - graphing examples in the real number system

An Elliptic Curve is the set of solutions to a Weierstrass Equation in the form:

y^2 = x^3 + Ax + B  (plus an extra point "infinity" that is the identity)

It must also be "non-singular", that is 4 A^3 + 27 B^2 must not equal 0

Below we will validate that curves are non-singular.

We will see various graphs of Elliptic Curves over the real numbers.

Elliptic Curves over the real numbers can be an ellipse and a curve as two separate graphs or a single curve.  Please note that the style does not influence the properties, as long as they meet the Weierstrass Equation and are non-singular.  Below the graphs will take us through various examples of this until we end up with the textbook prototypical elliptive curve.

```python
from sage.all import *
```

```python
def my_format_plus_minus(x):
    "given an integer, format it with a plus or minus, a blank, and the integer"
    
    if x < 0:
        s = "- " + str(abs(x))
    else:
        s = "+ " + str(abs(x))
    
    return s
```

#### Examples of validating an Elliptic Curve to be non-singular.  Examples below include those that are singular and those that are non-singular so you can see the difference.

```python
def my_validate_elliptic_curve_non_singular(A, B):
    "given A and B, validate that the elliptic curve will be non-singular, that is 4 A^3 + 27 B^2 not equal 0"
    
    A_string = my_format_plus_minus(A)

    B_string = my_format_plus_minus(B)
    
    print ("\nVerifying the Elliptic Curve to ensure it's non-singular:  y^2 = x^3 " + A_string + "x " + B_string + "\n")
    
    t = 4 * (A ** 3) + 27 * (B ** 2)
    
    print ("4 A^3 + 27 B^2 = 4 * (" + str(A) + "^3) + 27 * (" + str(B) + "^2) = " + str(t) + "\n")
    
    if t == 0:
        print ("Curve is singular - NOT acceptable")
    else:
        print ("Curve is non-singular - acceptable")
```

```python
A = 0
B = 0

my_validate_elliptic_curve_non_singular(A, B)
```

#### Examples of graphing various Elliptic Curves over the real numbers


We will see a progression.  It will start with a distinct ellipse on the left and a distinct curve on the right.  The ellipse and curve will move closer together.  Eventually, we will see the ellipse and curve as a single curve.  Finally, we will see an example of the style of the prototypical elliptic curve found in textbooks and published papers.

Please note that the style does not influence the properties, as long as they meet the Weierstrass Equation and are non-singular.

```python
def my_print_elliptic_curve_real_numbers(A, B, figure_size):
    "given A and B, which have been checked for non-singularity, print the elliptic curve over the real numbers"
        
    A_string = my_format_plus_minus(A)
    B_string = my_format_plus_minus(B)
    
    E = EllipticCurve([A,B])
    
    print ("\n=========================================================\n")
    print ("Elliptic Curve over the real numbers: \n\ny^2 = x^3 " + A_string + "x " + B_string + "\n")
    
    p = plot(E, gridlines='true', thickness=2)
    p.set_aspect_ratio(0.5)
    p.axes_labels(['$x$', '$y$'])
    
    show(p, figsize=figure_size)
```

```python
# this one appears as two distinct, unconnected graphs
# a small circular ellipse on the left
# a curve on the right

figure_size = 7

A = -5
B = -4

my_print_elliptic_curve_real_numbers(A, B, figure_size)
```

```python
# this one appears as two distinct, unconnected graphs
# a large ellipse on the left
# a curve on the right

figure_size = 7

A = -5
B = -1

my_print_elliptic_curve_real_numbers(A, B, figure_size)
```

```python
# this one appears as two distinct, unconnected graphs
# a large ellipse on the left
# a curve on the right

# you can see these coming closer to each other compared with the last graph

figure_size = 7

A = -5
B = 4

my_print_elliptic_curve_real_numbers(A, B, figure_size)
```

```python
# this one appears as a single graph

# comparing to the previous graph, you can see how the ellipse and curve have come together

figure_size = 7

A = -5
B = 5

my_print_elliptic_curve_real_numbers(A, B, figure_size)
```

```python
# this one is the prototypical elliptic curve shape found in textbooks and published papers

figure_size = 7

A = -5
B = 10

my_print_elliptic_curve_real_numbers(A, B, figure_size)
```
